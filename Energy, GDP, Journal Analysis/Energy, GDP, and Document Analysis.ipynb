```python
GDP
```

```python
ScimEn = pd.read_excel('scimagojr.xlsx')
```

```python
df = pd.merge(energy, GDP, how="inner", on="Country")
```

```python
final_df = pd.merge(df, ScimEn, how="inner", on="Country")
```

```python
final_df.set_index('Country', inplace = True)
```

```python
final_df = final_df[['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index', 
        'Energy Supply', 'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', 
        '2013', '2014', '2015']]
```

```python
final_df = final_df[final_df['Rank'] <= 15]
```

```python
final_df.sort('Rank', inplace = True)
```

```python
final_df
```

```python
def union_and_intersection():
    union_of_energy_and_gdp = pd.merge(energy, GDP, how="outer", on="Country")
    union_of_all_dataframes = pd.merge(union_of_energy_and_gdp, ScimEn, how="outer", on="Country")
    intersect_of_energy_and_gdp = pd.merge(energy, GDP, how="inner", on="Country")
    intersect_of_all_dataframes = pd.merge(union_of_energy_and_gdp, ScimEn, how="inner", on="Country")
    return len(union_of_all_dataframes) - len(intersect_of_all_dataframes)

union_and_intersection()
```

```python
def avg_gdp():
    years = ['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    return (final_df[years].mean(axis = 1)).sort_values(ascending = False).rename('avgGDP')

avg_gdp()
```

```python
years = ['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
final_df['Average GDP'] = final_df[years].mean(axis = 1)
```

```python
final_df
```

```python
def change_in_gdp():
    final_df['Average GDP'] = avg_gdp()
    final_df.sort_values(by = 'Average GDP', inplace = True, ascending = False)
    return abs(final_df.iloc[5]['2015']-final_df.iloc[5]['2006'])

change_in_gdp()
```
